# üóÇÔ∏è Python Data Structure - List

---

## üìñ Definition:
A **list** is an ordered, mutable (changeable) collection of items in Python. It can store elements of any data type ‚Äî integers, strings, floats, or even other lists.

‚úÖ Ordered: Items are stored in a specific sequence and maintain that order.

‚úÖ Mutable: You can add, remove, or change elements after the list is created.

‚úÖ Heterogeneous: A single list can hold different data types (e.g., int, str, float).

‚úÖ Iterable: Lists can be looped through using `for` or `while`.

‚úÖ Dynamic: You don‚Äôt need to declare the size of a list beforehand.


```python
# ‚úÖ Creating Lists

fruits = ["apple", "banana", "cherry"]  # List of strings
info = ["Hanif", 21, True, 3.14]        # Mixed data types
empty = []                              # An empty list

print(fruits)
print(info)
print(empty)

# ‚úÖ Accessing List Elements

print(fruits[0])    # Output: apple (1st item)
print(fruits[2])    # Output: cherry (3rd item)
print(fruits[-1])   # Output: cherry (last item)

# ‚úÖ Modifying Lists

fruits[1] = "mango"            # Change 'banana' to 'mango'
fruits.append("orange")        # Add 'orange' at the end
fruits.insert(1, "kiwi")       # Insert 'kiwi' at index 1

print(fruits)  # Output: ['apple', 'kiwi', 'mango', 'cherry', 'orange']

# ‚úÖ Removing Items from List

fruits.remove("apple")         # Remove 'apple' by value
del fruits[0]                  # Delete the item at index 0 (which is now 'kiwi')
fruits.clear()                 # Remove all items from the list

print(fruits)  # Output: []

# ‚úÖ Looping Through a List

colors = ["red", "green", "blue"]

for color in colors:
    print("Color:", color)

# ‚úÖ Useful List Functions

numbers = [5, 2, 9, 1, 5]

print("Length:", len(numbers))     # Output: 5
print("Minimum:", min(numbers))    # Output: 1
print("Maximum:", max(numbers))    # Output: 9
print("Sum:", sum(numbers))        # Output: 22

numbers.sort()                     # Sort list in ascending order
print("Sorted:", numbers)          # Output: [1, 2, 5, 5, 9]

numbers.reverse()                  # Reverse the list
print("Reversed:", numbers)        # Output: [9, 5, 5, 2, 1]

# ‚úÖ Practice Exercise

languages = ["Python", "JavaScript", "C++"]

for lang in languages:
    print("I like", lang)
```

# üî∑ Tuples in Python

## üìñ Definition:
A **tuple** is an ordered and immutable collection of items in Python. Like lists, tuples can store multiple data types, but **once created, their elements cannot be changed**.

---

## üîç Characteristics of Tuples:

‚úÖ **Ordered**: Elements have a fixed position and can be accessed using indexes.  
‚úÖ **Immutable**: Once a tuple is defined, you cannot modify, add, or remove its items.  
‚úÖ **Heterogeneous**: Can contain elements of different types (int, str, float, etc.).  
‚úÖ **Iterable**: Can be looped through using `for` loops.  
‚úÖ **Faster than lists**: Since tuples are immutable, Python can optimize their storage.

---

## üß† Use Cases:
- Data that should not change (e.g., coordinates, RGB values, database records).
- Used as keys in dictionaries (only immutable types can be keys).

```python
# ‚úÖ Creating Tuples

person = ("Hanif", 21, "Pakistan")
coordinates = (34.5, 71.3)
single_item = ("Python",)   # Note the comma! Without it, it's just a string

print(person)
print(coordinates)
print(single_item)

# ‚úÖ Accessing Tuple Elements

print(person[0])      # Output: Hanif
print(person[-1])     # Output: Pakistan

# ‚úÖ Immutability Check

# person[0] = "Ali"   # ‚ùå This will cause an error: 'tuple' object does not support item assignment

# ‚úÖ Looping Through a Tuple

for item in person:
    print("Item:", item)

# ‚úÖ Tuple Methods

numbers = (5, 3, 7, 3, 9)

print("Count of 3:", numbers.count(3))  # Counts how many times 3 appears
print("Index of 7:", numbers.index(7))  # Finds the index of the first occurrence of 7

# ‚úÖ Tuple Unpacking

name, age, country = person
print("Name:", name)
print("Age:", age)
print("Country:", country)
```

# üóÉÔ∏è Dictionaries in Python

## üìñ Definition:
A **dictionary** is an **unordered, mutable** collection of **key-value pairs**. It allows you to store and retrieve data using meaningful keys instead of index positions.

---

## üîç Characteristics of Dictionaries:

‚úÖ **Key-value pairs**: Each item is stored as a pair ‚Äî `key: value`.  
‚úÖ **Unordered** (before Python 3.7): The order of items is not guaranteed.  
‚úÖ **Mutable**: You can add, remove, or change items.  
‚úÖ **Unique keys**: Keys must be unique and immutable (e.g., strings, numbers, tuples).  
‚úÖ **Dynamic**: You can grow or shrink a dictionary during runtime.

---

## üß† Use Cases:
- Storing structured data (e.g., user profiles, JSON responses).
- Efficient lookups and fast data access using keys.
- Django uses dictionaries for settings, forms, request/response data, etc.


```python
# ‚úÖ Creating a Dictionary

student = {
    "name": "Hanif",
    "age": 21,
    "course": "Python for AI",
    "is_active": True
}

print(student)

# ‚úÖ Accessing Values

print(student["name"])       # Output: Hanif
print(student.get("age"))    # Safer way to access keys

# ‚úÖ Modifying Values

student["age"] = 22          # Change age
student["country"] = "Pakistan"  # Add new key-value pair

print(student)

# ‚úÖ Removing Items

del student["is_active"]     # Remove a key
print(student)

# ‚úÖ Looping Through a Dictionary

for key in student:
    print(f"{key} => {student[key]}")

# ‚úÖ Useful Dictionary Methods

print("Keys:", student.keys())        # All keys
print("Values:", student.values())    # All values
print("Items:", student.items())      # All key-value pairs

# ‚úÖ Dictionary Nesting (Advanced Use)

user = {
    "username": "hanifullah",
    "profile": {
        "email": "hanif@example.com",
        "location": "Pakistan"
    }
}

print(user["profile"]["email"])  # Accessing nested dictionary
```

# üß© Python Set

## üìñ Definition:
A **set** is an **unordered**, **mutable**, and **unindexed** collection in Python that **does not allow duplicate values**. Sets are useful for storing **unique items** and performing operations like **union**, **intersection**, and **difference**.

Sets are created using **curly braces `{}`** or the `set()` function.

---

## üîç Characteristics of Sets:

‚úÖ **Unordered**: The items do not maintain any specific order.  
‚úÖ **Unindexed**: You cannot access items using an index like in lists or tuples.  
‚úÖ **Mutable**: You can add or remove items (but the items themselves must be immutable types).  
‚úÖ **No Duplicates**: Sets automatically remove duplicate elements.  
‚úÖ **Supports Set Operations**: Such as union, intersection, and difference ‚Äî just like in mathematics.

---

## üß† Use Cases:
- Removing duplicate values from a list.
- Performing fast membership tests (`in`).
- Mathematical operations like finding common or unique elements between collections.


```python
# ‚úÖ Creating Sets

numbers = {1, 2, 3, 4, 4, 5}
print(numbers)  # Output: {1, 2, 3, 4, 5} ‚Äî duplicates are removed

empty_set = set()  # Correct way to create an empty set
print(type(empty_set))  # Output: <class 'set'>

# ‚úÖ Adding and Removing Elements

numbers.add(6)       # Add one item
numbers.update([7, 8])  # Add multiple items
numbers.discard(3)   # Remove item safely (no error if not found)
numbers.remove(2)    # Remove item (throws error if not found)

print(numbers)

# ‚úÖ Set Operations

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print("Union:", a | b)           # {1, 2, 3, 4, 5, 6}
print("Intersection:", a & b)    # {3, 4}
print("Difference (a - b):", a - b)  # {1, 2}
print("Symmetric Difference:", a ^ b)  # {1, 2, 5, 6}

# ‚úÖ Membership Test

print(5 in a)   # False
print(4 in a)   # True
```
